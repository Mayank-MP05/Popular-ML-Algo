```
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#Datasets Importing
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```
test.head()
```

```
# Dropping the Expected output column in Training Data
train_x = train.drop(columns=['Item_Outlet_Sales'],axis=1)

# Actual training Data Output
train_y = train['Item_Outlet_Sales']

train_x.head()
```

```
# Same Separation for Testing Data
test_x = test.drop(columns=['Item_Outlet_Sales'],axis=1)
test_y = test['Item_Outlet_Sales']
```

```
#creating model
model = LinearRegression()

# fit the model with the training data
model.fit(train_x,train_y)
```

```
print(model.__dict__)
```

```
# y = mx + c

# m = coefficeints of the trained model
print('\nCoefficient of model :', model.coef_)

# c = intercept of the model
print('\nIntercept of model',model.intercept_)
```

```
# model prediction
# predict the target on the test dataset
predict_train = model.predict(train_x)
print('\nItem_Outlet_Sales on training data',predict_train) 
```

```
# Root Mean Squared Error on training dataset
rmse_train = mean_squared_error(train_y,predict_train)**(0.5)
# print('\nRMSE on train dataset : ', rmse_train)

# predict the target on the testing dataset
predict_test = model.predict(test_x)
# print('\nItem_Outlet_Sales on test data',predict_test) 

# Root Mean Squared Error on testing dataset
rmse_test = mean_squared_error(test_y,predict_test)**(0.5)
# print('\nRMSE on test dataset : ', rmse_test)
import numpy as np
from sklearn.metrics import accuracy_score
# Checking for Accuracy Score
# test_y1 = pd.Series.to_numpy(test_y)
print(type(test_y))

# acc = accuracy_score(test_y,test_y)
# print("Accuracy of the model is : "+acc)
```
